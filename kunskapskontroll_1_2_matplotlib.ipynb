### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

There is two ways: 
* OO-approach 
* matlab-style

The first one is recommended in general and especially for big and complicated projects and the second one is used for smal and quick plotting. 

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

* Figure is a "painting". It containes all the elements of plot.
* Axes is a "canvas", an area where we plot.
* Axis is x- and y- axis. 
* Artist is every element on Figure, our "painting". 


### When plotting in Matplotlib, what is the expected input data type?

It depends on the type of plot, but we can generally say that it's numeric values or data structures like NumPy arrays or lists that contain numeric values. The most common input type is the NumPy array.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OO-approach 

x = np.linspace(-4,4, 1000)

fig, ax = plt.subplots()

ax.plot(x, x**2)
ax.set_title("OO-approach")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()

# MATLAB-approach

plt.plot(x, x**2)
plt.xlabel('x')
plt.ylabel("y")
plt.title("MATLAB-approach")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Scatterplot
fig,ax = plt.subplots(3,1,gridspec_kw={'hspace':0.8})

ax[0].scatter(random_data_x, random_data_y, edgecolor='g', color='blue') #Ingen aning varfr punkterna r rda...
ax[0].set_title('Scatter plot')
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")


# Bar Plot

ax[1].bar(names, values, color=['red', 'green', 'blue', 'purple', "pink"])

ax[1].set_title('Bar plot')
ax[1].set_ylabel("Quantity")

# Lineplot

ax[2].plot(x, y, color='red')
ax[2].set_title('Line plot')
ax[2].set_xlabel("x")
ax[2].set_ylabel("y")

ax[0].scatter(random_data_x, random_data_y, alpha=0.2, color='red')

plt.show()
```
